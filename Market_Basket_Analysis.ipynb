```python
###Importing the dataset####
import pandas as pd
df=pd.read_csv("C:/Users/shreyas/Downloads/shreyas_data/Groceries_dataset.csv")
df
```

```python
import matplotlib.pyplot as plt
x=df['itemDescription'].head(10)
y=df['itemDescription'].value_counts().head(10)
plt.bar(x,y)
plt.show()
```

```python
dummies=pd.get_dummies(df['itemDescription'])
df2=df.join(dummies)
```

```python
items=df['itemDescription'].unique()
df3=df2.groupby(['Member_number','Date'])[items[:]].sum()
```

```python
df4=df3.reset_index()
```

```python
df5=df4.drop(['Member_number','Date'],axis=1)
```

```python
df5
```

```python
for i in range(len(df5)):
    for j in (df5.columns):
        if df5.loc[i,j]>0:
            df5.loc[i,j]=j
df5.head(10)
```

```python
transactions=[]
for i in range(len(df5)):
    x=[]
    for j in (df5.columns):
        if df5.loc[i,j]!=0:
            x.append(df5.loc[i,j])
    transactions.append(x)
transactions
```

```python
from apyori import apriori
results = apriori(transactions,min_support=0.0003,min_confidence=0.02,min_lift=3,min_length=2,target='rules')
```

```python
results = list(results)
```
